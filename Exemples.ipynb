Probabilité de retour :

Avec une modélisation par une distribution de Poisson on obtient :

$$P_t(r)=\frac{(\mu t)^r}{r!}e^{-\mu t}=\frac{(t/T)^r}{r!}e^{-t/T}$$

Avec
- $t$ le nombre de periode,
- $r$ le nombre d'occurence du phénomène
- $T$ la periode de retour
- $\mu = \frac 1 T$

**Exemple :** $P_{10}(1)$ est la probabilité d'avoir $1$ occurence sur une periode de $10$ ans

On déduit alors la probabilité d'avoir au moins $k$ occurences sur un nombre de periode $t$ : $$\forall k \in \mathbb{N},P_t(r\geq k) = 1 - P_t(r<k) = 1 - \sum_{i=0}^{k-1} P_t(i)$$
